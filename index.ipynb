# Importações Gerais


```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('https://raw.githubusercontent.com/ThiagoJaime/report-py/main/dados.csv')
```

# Compreensão dos Dados:

```python
# Primeiras 10 linhas
df.head(10)
```

```python
# Informações Gerais
df.info()
```

```python
# Estatísticas descritivas
df.describe()
```

# Análises


**Total de Horas Trabalhadas**

```python
totHoras = df['Horas Trabalhadas'].sum()

totHoras
```

**Média Diária de Horas Trabalhadas**

```python
mediaHora = df['Horas Trabalhadas'].mean()

mediaHora.round(2)
```

**Total de Bugs Corrigidos**

```python
totBugs = df['Bugs Corrigidos'].sum()

totBugs
```

**Média Diária de Bugs Corrigidos**


```python
mediaBugs = df['Bugs Corrigidos'].mean()

mediaBugs.round(2)
```

**Total de Tarefas Concluídas**

```python
totTaf = df['Tarefas Concluídas'].sum()

totTaf
```

**Média Diária de Tarefas Concluídas**

```python
mediaTaf = df['Tarefas Concluídas'].mean()

mediaTaf.round(2)
```

**Produtividade Diária**

```python
df['Produtividade Diária'] = df['Tarefas Concluídas'] / df['Horas Trabalhadas']

df[['Dia', 'Produtividade Diária']]
```

```python
df['Horas Trabalhadas'].plot(kind='bar', figsize=(12, 6), title='Horas Trabalhadas por Dia')
plt.xlabel('Dia')
plt.ylabel('Horas Trabalhadas')
plt.show()
```

```python
df['Bugs Corrigidos'].plot(kind='line', marker='o', figsize=(14, 6), title='Bugs Corrigidos por Dia')
plt.xlabel('Dia')
plt.ylabel('Bugs Corrigidos')
plt.grid(True)
plt.show()
```

```python
df['Tarefas Concluídas'].plot(kind='bar', figsize=(12, 6), title='Tarefas Concluídas por Dia')
plt.xlabel('Dia')
plt.ylabel('Tarefas Concluídas')
plt.show()
```

```python
df['Produtividade Diária'].plot(kind='line', marker='o', figsize=(12, 6), title='Produtividade Diária')
plt.xlabel('Dia')
plt.ylabel('Produtividade Diária (Tarefas/Hora)')
plt.grid(True)
plt.show()
```

# Insights

### Distribuição de Horas Trabalhadas:

- Manter uma média constante de horas trabalhadas em torno de 6-7 horas por dia parece ser eficaz para manter a produtividade em termos de bugs corrigidos e tarefas concluídas.

### Dias da Semana com Maior Produtividade:

- **Segunda e Sexta-feira** são dias de alta produtividade. Considerar distribuir tarefas críticas nesses dias pode ser benéfico.
- Finais de semana têm uma produtividade menor, o que pode ser considerado para planejamento de atividades menos críticas ou descanso.

### Balanceamento de Tarefas:

- A média de tarefas concluídas por dia é bastante consistente, o que indica um bom balanceamento de trabalho. Continuar monitorando para evitar sobrecarga em dias específicos.
